```python
def exponential_smoothing(series, alpha):
    result = [series[0]] # first value is same as series
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    return result
```

```python
# сделаем примеры и модель 
def split_data( data, split_date ):
    return data.loc[data.index.get_level_values('Date') <= split_date].copy(), \
           data.loc[data.index.get_level_values('Date') >  split_date].copy()
```

```python
train, test = split_data(df, '2014-01-01')

X_train = train.iloc[:-1,:]
y_train = train[df.columns[0]].values[1:]
X_test = test.iloc[:-1,:]
y_test = test[df.columns[0]].values[1:]

plt.figure(figsize=(15,7))
plt.xlabel('Время')
plt.ylabel('МВт')
plt.plot(train.index.get_level_values('Date'),train[df.columns[0]], label='train data')
plt.plot(test.index.get_level_values('Date'),test[df.columns[0]], label='test data')
plt.title('Тестовые и тренировочные данные')
plt.legend()
plt.show()
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
model_gb = GradientBoostingRegressor(max_depth=15, random_state=0, n_estimators=100)
model_gb.fit( X_train, y_train ) 
X_test_pred_gb = model_gb.predict(X_test)

plt.figure( figsize=( 20, 4 ))

plt.xlabel('Время')
plt.ylabel('МВт')
plt.plot(test.index[1:],X_test_pred_gb, label='predict GB data' )
plt.plot(test.index[1:],y_test, label='test data')
plt.plot(d_data.loc['2014-01-02':].index, d_data.loc['2014-01-02':]['PJME_MW'], label='исходный ряд')

plt.legend()
plt.show()

er_g = mean_absolute_percentage_error(y_true=y_test, y_pred=X_test_pred_gb)

print('ошибка градиентного бустинга :  ', er_g, '%')
```

Простыми словами:
Стационарность ряда - это свойство ряда не менять своих статистических характеристик с течением времени, т.е. постоянство матожидания, дисперсии и независимость ковариационной функции от времени (должна зависеть только от расстояния между наблюдениями).

Тренд — основная тенденция изменения чего-либо.

Сезонность - явление, характеризующееся периодическими колебаниями, которые повторяются каждый сезон по предсказуемой схеме. Причем, под словом "сезон" не обязательно подразумевается смена времен года (лето, осень, вестна, зима). Сезоном может быть, например, сезон чемпионата страны по футболу, сезон отпусков, сезон выплаты годовых премий, фазы Луны (новолуние, полнолуние, восходящая луна) и т.д.

Таким образом:

a) Ряд не стационарный. Есть сезоны. Тренд отсутсвует (если не считать, что в целом температура на Земле имеет тренд к повышению по причине глобального потепления).

б) Ряд не стационарный. Есть сезоны (например, смена времен года => смена ассортимента приобретаемых товаров => смена кол-ва потраченных денег). Тренд имеется (растет уровень жизни на протяжении всей человеческой истории => больше трат. Плюс инфляция).

в) Ряд не стационарный. Есть сезоны (набор подкожного жира и шерсти на коже в холодное время года => рост массы и, соответственно, сокращение жира и шерсти в теплое время года). Тренд имеется (котенок растет).

г) Ряд не стационарный. Есть сезоны (зимой рост, как и прочие процессы, замедляется по причине холодной температуры - закон Физики). Тренд имеется (объем деревьев растет).
